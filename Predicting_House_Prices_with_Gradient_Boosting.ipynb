# Predicting House Prices with Gradient Boosting

In this project, we utilize gradient boosting,to predict house prices.

```python
# Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics  import mean_absolute_error, r2_score

```

## Loading In DataSet

```python
# Read in Data from CSV
df = pd.read_csv(r"/content/MELBOURNE_HOUSE_PRICES_LESS.csv")
df.head()
```

## Data Cleaning

```python
# Delete Unneded Columns
drop_columns = ["Address", "Method", "SellerG", "Date", "Postcode", "Regionname", "Propertycount"]
df.drop(columns = drop_columns, inplace = True)
df.head()
```

```python
# Remove rows with missing values
df.dropna(axis=0, how="any", subset=None, inplace=True)
# Convert non-numerical data using one hot encoding
df = pd.get_dummies(df,columns = ["Suburb", "CouncilArea", "Type"])
df.head()
```

## Training and Test Data

```python

# Assign dependent and independent variables
X = df.drop("Price", axis = 1)
Y = df["Price"]

# Split data into test/train set (70/30) split and shuffle
x_train , x_test, y_train, y_test = train_test_split(X,Y, test_size =0.30, shuffle=True )

```

## Building Model

```python
# Deffine Algorithem
model = ensemble.GradientBoostingRegressor(
    n_estimators= 150, # Number of Trees
    learning_rate = 0.1, # Influence of each tree
    max_depth = 5, # Maxinum number of layers
    min_samples_split = 4, # Minimum number of samples
    min_samples_leaf=6, # Minimum number of samples needed before split
    max_features=0.6,
    loss = "huber"
)

# Run Model on training data
model.fit(x_train, y_train)

```

## Model Performance Metrics

```python
# Check Model Accuracy
mae_train= mean_absolute_error(y_train,model.predict(x_train))
print("Training Set Mean Abs Error %.2f" % mae_train)

mae_test = mean_absolute_error(y_test,model.predict(x_test))
print("Test Set Mean Abs Error %.2f" % mae_test)
```
